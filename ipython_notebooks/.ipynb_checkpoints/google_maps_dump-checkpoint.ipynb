```python
from dotenv import load_dotenv
import os

load_dotenv()

API_KEY=os.getenv("GOOGLEMAPS_API_KEY")
```

```python
import googlemaps

gmaps = googlemaps.Client(key=API_KEY)
north_york_ctr_lat_lng = (43.76921917771091, -79.41288966236526)

# Google Maps API documentation is here: https://developers.google.com/maps/documentation/places/web-service/search
# Python API documentation: https://github.com/googlemaps/google-maps-services-python/tree/master
```

```python
# Given a list of restaurant names, get details for each restaurant

restaurants_google_maps_json = []

# read OSM json file
import json
f = open('../src/assets/restaurants.json', encoding="utf8")
data = json.load(f)
f.close()

# call google maps API
counter = 0
for feature in data["features"]:
    if counter == 2:
        break
    counter += 1
    print(counter)
    restaurants_google_maps_json.append(gmaps.find_place(feature['properties']['name'],
                                                         "textquery",
                                                         fields=["name","formatted_address","types","rating","user_ratings_total","price_level",\
                                                                 "opening_hours","business_status","photos","place_id","permanently_closed"],
                                                         location_bias="circle:2000@" + str(north_york_ctr_lat_lng[0]) + "," + str(north_york_ctr_lat_lng[1]), 
                                                         language='en-US'))

# For gmaps.find_place: Valid values for the `fields` param for `find_place` are 
# 'opening_hours', 'rating', 'geometry/viewport/southwest/lng', 'business_status', 
# 'photos', 'geometry/location/lng', 'geometry/viewport/southwest/lat', 'price_level', 
# 'geometry/viewport/northeast', 'types', 'plus_code', 'user_ratings_total', 'geometry/location',
# 'geometry/viewport/northeast/lng', 'geometry/viewport', 'geometry/location/lat', 'place_id',
# 'geometry', 'icon', 'geometry/viewport/northeast/lat', 'geometry/viewport/southwest', 'formatted_address',
# 'permanently_closed', 'name'

# Turn into readable JSON
restaurants_google_maps_json = json.dumps(restaurants_google_maps_json, indent=2)
```

```python
# Get photos, given a google maps "photo id"
# can get photo id thru find_place calls

import json, pprint
f = open('src/assets/googlemaps.json', encoding="utf8")
data = json.load(f)
f.close()

pprint.pprint(data)
```

```python
counter = 0

for restaurant in data:
    if restaurant['status'] == 'OK':
        try:
            photo_reference = restaurant['candidates'][0]['photos'][0]['photo_reference']
            photo = gmaps.places_photo(photo_reference,max_width=1600)

            # write photo
            with open("src/assets/gmaps_photos/" + photo_reference + ".jpg", "wb") as fp:
                for chunk in photo:
                    fp.write(chunk)

            counter += 1
            print(counter)
        except KeyError:
            pass

        
print(counter)
```
